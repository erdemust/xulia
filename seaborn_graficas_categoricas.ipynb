<h1 style="color:purple;">Datos Categóricos a Visualizar</h1> 

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="ticks", context="talk", palette="tab10")

consumos = sns.load_dataset("tips")

consumos.columns = ["total", "propina", "genero", "fumador",
                    "dia", "tipo", "comensales"]

consumos
```

<br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br>
<h1 style="color:darkblue;"> Gráficas categóricas de Dispersión</h1>
<h2 style="color:indianred;"> Gráfica stripplot </h2>

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="strip", hue="comensales", s=7, col="fumador")

sns.catplot(x="dia", y="total", data=consumos,
            kind="strip", jitter=False, hue="comensales", s=7, col="fumador")

plt.show()
```

<h2 style="color:indianred;"> Gráfica swarmplot </h2>

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="swarm", hue="comensales", s=7, col="fumador")
plt.show()
```

<br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br>
<h1 style="color:darkblue;"> Gráficas categóricas de Distribución</h1>
<h2 style="color:indianred;"> Gráfica boxplot</h2>

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="box", col="fumador")
plt.show()
```

<h2 style="color:indianred;"> Gráfica boxenplot</h2>

```python
sns.catplot(x="dia", y="total", data=consumos, 
            kind="boxen", col="fumador")
plt.show()
```

<h2 style="color:indianred;"> Gráfica violinplot</h2>

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="violin")

sns.catplot(x="dia", y="total", data=consumos,
            kind="violin", hue="fumador")

sns.catplot(x="dia", y="total", data=consumos,
            kind="violin", hue="fumador", split=True)

plt.show()
```

<br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br><br>
<h1 style="color:darkblue;"> Gráficas categóricas de Estimación</h1>
<h2 style="color:indianred;"> Gráfica pointplot</h2>

```python
sns.catplot(x="dia", y="total", data=consumos, 
            kind="point", col="fumador")
plt.show()
```

<h2 style="color:indianred;"> Gráfica barplot</h2>

```python
sns.catplot(x="dia", y="total", data=consumos,
            kind="bar", hue="comensales")
plt.show()
```

<h2 style="color:indianred;"> Gráfica countplot </h2>

```python
sns.catplot(x="dia", data=consumos,
            kind="count", hue="comensales")
plt.show()
```
